# Load Model

```python
import pandas as pd
import joblib
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("data_cleaned.csv", header=0)
data
```

```python
train = data.sample(n=1800,random_state=175)
test = data.drop(train.index)

# Assuming your DataFrame is named 'train' and the target column is named 'target'
X = train.drop('litho', axis=1).values
y = train['litho'].values.reshape(-1, 1)
y = y.ravel()


X_test = test.drop('litho', axis=1).values
Y_test = test['litho'].values.reshape(-1, 1)
Y_test = Y_test.ravel()
```

```python
model_file = "litho.joblib"
model = joblib.load(model_file)

y_pred = model.predict(X)
accuracy = accuracy_score(y, y_pred)
print("Accuracy:", accuracy)
```
